## Introduction

This tutorial will provide you a breif introduction of visualizing the similarities between materials using t-Distributed Stochastic Neighbor Embedding (t-SNE). There are myraid types of materials known to human beings and each of them has hundreds of features, including microscopic structure (topology), material type, etc. With the huge number of features, it is very hard for researchers to directly visualize the similarities between different kinds of materials because of the high dimensionality. In materials discovering process, when a new materials is synthesized or simulated, it will be very helpful for researchers to have a tool that can show them if this new material is similar to any kind of already-known materials. 

t-SNE is developed as a non-linear dimensionality reduction technique which is very efficient in visualizing high dimensional data in 2-d space. When projected to a 2-d space by t-SNE, similar high dimensional data points will be modeled closer to each other than dissimilar data points. This amazing property makes t-SNE a ideal tool for visualizing known materials and potentially predicting the similarity between newly discovered material and existing materials.

## Loading data and visualizing material structure

Now all required libraries are installed, time to load the data files. PDB files in the pdbs folder are named with a specific pattern for the convenience of users. In the pdbs folder, there are generally 5 types of materials:
1. Graphene (file named as *.gra.pdb, * is the ID of each one)
1. Weird-shape Graphene (Slightly different compared with Graphene, named as *.shape.pdb, * is the shape of the material)
1. Phosphorene (file named as *.phos.pdb, * is the ID of each one)
1. MoS$_{2}$ (file named as *.mos2.pdb, * is the ID of each one)
1. MOF (file named as *.mof.pdb, * is the metal atom type of each one)

When loading the data, we will firstly load all files in one type into a list of file name strings. We will also store the number of materials in each type for future use. Then with the pdb2mol method, we can use rdkit to read the pdb files into rdkit.Chem.rdchem.Mol objects (Mol objects for abbreviation). Mol object is a built-in object in RDKit pacakge that can be visualized to check if the structure of materials are reasonable. Mol object contains topology and chemical infomation of material and will later be featurized for t-SNE use.

```python
gra_files = glob.glob("pdbs/*.gra.pdb")
shape_files = glob.glob("pdbs/*.shape.pdb")
phos_files = glob.glob("pdbs/*.phos.pdb")
mos2_files = glob.glob("pdbs/*.mos2.pdb")
mof_files = glob.glob("pdbs/*.mof.pdb")
n_gra = len(gra_files)
n_shape = len(shape_files)
n_phos = len(phos_files)
n_mos2 = len(mos2_files)
n_mof = len(mof_files)
```

```python
def pdb2mol(files):
    mol = []
    for i in range(len(files)):
        m = Chem.MolFromPDBFile(files[i])
        mol.append(m)
    return mol
```

```python
gra_mol = pdb2mol(gra_files)
shape_mol = pdb2mol(shape_files)
phos_mol = pdb2mol(phos_files)
mos2_mol = pdb2mol(mos2_files)
mof_mol = pdb2mol(mof_files)
```

```python
# Tiny check if all files in gra_mol are rdkit.Chem.rdchem.Mol objects
print(all([isinstance(x, Chem.rdchem.Mol) for x in gra_mol]))
```

After all pdb files are loaded into Mol objects, we can simply check the structure of all materials:

```python
check_mol = [gra_mol[0], shape_mol[0], phos_mol[0], mos2_mol[0], mof_mol[0]]
label = ['Graphene', 'Weird-shape Graphene', 'Phosphorene', 'MoS2', 'MOF']
Draw.MolsToGridImage(check_mol,molsPerRow=3,subImgSize=(200,200),legends=[x for x in label])
```

Looks like all materials are successfully loaded!

## Material featurization

Now, we have the information of each material as Mol objects. How can we extract the features into a binary numpy array?
Each material is consisted of many atoms, and the features of each atom (electron donor/acceptor, aromatic, halogen, acidity, etc) combine together to determine the properties of the material. When featurizing a specific material, what we do is to get a series of features of each atoms in that material. This series of features is called Morgan Fingerprints.

Each material has its own unique Morgan Fingerprints. Therefore, the process of featurization of materials consists of two steps:
1. Convert the Mol objects to Morgan Fingerprints objects
    - When doing this, we need to supply the program with two arbituary values: cutoff distance and number of features. Cutoff distance `cutoff` is the radius (with Angstrom as unit) that defines a spherical region center at the location of each atom. It tells the program that we do not want any information outside the spherical region of each atom because they are too far away. Usually people use `cutoff = 2` to take only most vicinal features, we use `cutoff = 4` to extract more data for better t-SNE result. The number of features `n_features` defines how many features you want to extract for each atoms. Conventionally people use `n_features = 512` or `1024`. However, since the numbers of types of features are almost countless, to avoid sparsity in output array (increase uniqueness of material), we use `n_features = 128`.
       <br/><br/>
1. Get Numpy array representation of Morgan Fingerprints for t-SNE use

```python
# Define the constant parameters
n_features = 128
cutoff_dist = 4
```

```python
def featurization(mol, cutoff, features):
    # Step 1: Convert Mol object to Morgan Fingerprints
    fps = []
    for m in mol:
        fps.append(AllChem.GetMorganFingerprintAsBitVect(m, 
                        cutoff, useFeatures=True, nBits = features))
    # Step 2: Convert Morgan Fingerprints to numpy array
    matrices = []
    for i in range(len(fps)):
        array = np.zeros((1,));
        DataStructs.ConvertToNumpyArray(fps[i], array);
        matrices.append(array)
    matrices = np.asarray(matrices)
    return matrices
```

```python
gra_data = featurization(gra_mol, cutoff_dist, n_features)
shape_data = featurization(shape_mol, cutoff_dist, n_features) 
phos_data = featurization(phos_mol, cutoff_dist, n_features)
mos2_data = featurization(mos2_mol, cutoff_dist, n_features)
mof_data = featurization(mof_mol, cutoff_dist, n_features)
```

```python
all_data = np.concatenate((gra_data, shape_data, phos_data, mos2_data, mof_data))
```

## Using t-SNE to visulize similarities

We employed t-SNE from scikit-learn with a typical `perplexity` value as 10. The result we obtained is phenomenal. Data points within same material group are closer to each other while keeping distance to data points in other material groups. 

One interesting thing to notice is that although Weird-shaped Graphenes have different geometries comparing to other Graphenes, they are closer to Graphene comparing to other material groups. This shows that t-SNE can successfully recognize the material type similarity between Graphene and Weird-shape Graphene. At the mean time, t-SNE also seprate them because their topologies are different.

```python
def draw_tsne(data, perplexity = 10, show_legend = True):
    start_shape = n_gra
    start_phos = start_shape+n_shape
    start_mos2 = start_phos+n_phos
    #########################################
    # TSNE section
    tsne = TSNE(n_components = 2,perplexity = perplexity)
    points = tsne.fit_transform(data)
    #########################################
    # Plot section
    plt.scatter(points[:n_gra,0], points[:n_gra,1], 
                label = 'Graphne')
    plt.scatter(points[start_shape:start_phos,0], 
                points[start_shape:start_phos,1],
               label = 'Weird-shape Graphene')
    plt.scatter(points[start_phos:start_mos2,0], 
                points[start_phos:start_mos2,1],
               label = 'Phosphorene')
    plt.scatter(points[start_mos2:-n_mof,0], 
                points[start_mos2:-n_mof,1],
               label = r'MoS$_2$')
    plt.scatter(points[-n_mof:,0], points[-n_mof:,1],
               label = 'MOF')
    #########################################
    # Plot beautifying section
    plt.tick_params(direction = 'in',width = 1.5)
    plt.xticks([],[''])
    plt.yticks([],[''])
    if show_legend:
        plt.legend(bbox_to_anchor=(1.04,1),frameon = False, loc="upper left", fontsize =10)
```

```python
fig = plt.figure(figsize = [5, 4], dpi = 120)
draw_tsne(all_data)
plt.title('t-SNE, Perplexity = 10')
# Make the frame of plot thicker
ax = gca()
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
plt.tick_params(direction = 'in',width = 1.5)
plt.show()
```

## The influence of perplexity value to the performance of t-SNE

Perplexity is an important hyperparameter in t-SNE which can directly change the performance. An OVERSIMPLIFIED explanation of the function of perplexity is that it defines the number of related neighbors of each point. In previous step, we used t-SNE with perplexity = 10 for material similarity visualization. Let's take a look at the influence on the t-SNE performance by different perplexity and see if there is an optimal option.

```python
perp_list = [5, 10, 20, 30, 40, 50]
subplot_list = [231, 232, 233, 234, 235, 236]
fig = plt.figure(figsize = [8, 5], dpi = 120)
for i, subplot in enumerate(subplot_list):
    plt.subplot(subplot)
    if subplot == 233:
        draw_tsne(all_data, perplexity = perp_list[i])
    else:
        draw_tsne(all_data, perplexity = perp_list[i], show_legend = False)
    plt.title('Perplexity = %d' %perp_list[i])
    ax = gca()
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2)
```

From the plots, we can clearly see that perplexity have decisive influence on t-SNE performance. In total, we have 36 data points. When perplexity value increases, the boundary between material groups become unclear. When perplexity value is 30, which relatively high but not exceeding the number of total points, t-sne is still capable of distinguishing between materials, only with narrow gap between material groups. However, when perplexity value exceeds the total number of points (Perplexity = 40 or 50), t-SNE can only deliver a messy cluster of data points without any clear boundary between groups. Therefore, when tuning the t-SNE model for similarity visualization, we have to set the perplexity value to be smaller than total number of points.

Another intersting point to notice is that at very lower perplexity (perplexity = 5 or smaller), data points seem to form much gathered clusters. The differences between data points in a certain group tend to be ignored. Therefore, a very small perplexity value should also be avoided. 

## Comparing t-SNE with PCA

Principal Component Analysis (PCA) is a well-known tool used in classification and dimensionality reduction. Although PCA and t-SNE can sometimes be used for same purpose (such as in this tutorial), they are fundamentally different tools:
- t-SNE reduces dimensionality by project data points into 2-d plane while keeping the distance (similarity) between points the same.
- PCA reduces dimensionality by selecting pricipal component with most variations.

t-SNE is relatively newer tool compared with PCA, which is invented more than 100 years ago. Let's see if t-SNE is more advanced for this material similarity visualization project.

```python
def pca_result(data):
    start_shape = n_gra
    start_phos = start_shape+n_shape
    start_mos2 = start_phos+n_phos
#     fig = plt.figure(figsize = [6, 4], dpi = 200)
    #########################################
    # PCA section
    pca = PCA(n_components = 2)
    pca_data = pca.fit_transform(data)
    #########################################
    # Plot section
    plt.scatter(pca_data[:n_gra,0], pca_data[:n_gra,1],
               label = 'Graphne')
    plt.scatter(pca_data[start_shape:start_phos,0], 
                pca_data[start_shape:start_phos,1],
               label = 'Weird-shape Graphene')
    plt.scatter(pca_data[start_phos:start_mos2,0], 
                pca_data[start_phos:start_mos2,1],
               label = 'Phosphorene')
    plt.scatter(pca_data[start_mos2:-n_mof,0], 
                pca_data[start_mos2:-n_mof,1],
               label = r'MoS$_2$')
    plt.scatter(pca_data[-n_mof:,0], pca_data[-n_mof:,1],
                label = 'MOF')
    #########################################
    # Plot beautifying section
    ax = gca()
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2)
    plt.title('PCA')
    plt.tick_params(direction = 'in',width = 1.5)
    plt.xticks([],[''])
    plt.yticks([],[''])
    plt.legend(bbox_to_anchor=(1.04,1),frameon = False, loc="upper left", fontsize =10)
```

```python
fig = plt.figure(figsize = [9, 4], dpi = 100)

plt.subplot(121)
draw_tsne(all_data, show_legend = False)
plt.title('t-SNE, Perplexity = 10')
ax = gca()
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(2)
plt.tick_params(direction = 'in',width = 1.5)

plt.subplot(122)
pca_result(all_data)
plt.show()
```

As shown in the PCA plot, although materials are successfully seprated by PCA, PCA seems to fail to distinguish the difference inside each material group. For example, for the group of phosphorene and MOF, all data points are overlapping with each other. Generally speaking, PCA managed to classify materials by its type, but fails to distinguish materials by other features such as topology. Moreover, PCA does not have a adjustable parameter to control the accuracy of classification just like perplexity to t-SNE. In conclusion, t-SNE is still the more suitable tool for material similarity visualization.
