```python
import glob
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem, Draw
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from pylab import *
```

```python
n_features = 128
cutoff_dist = 4
```

```python
gra_files = glob.glob("pdbs/*.gra.pdb")
shape_files = glob.glob("pdbs/*.shape.pdb")
phos_files = glob.glob("pdbs/*.phos.pdb")
mos2_files = glob.glob("pdbs/*.mos2.pdb")
mof_files = glob.glob("pdbs/*.mof.pdb")
n_gra = len(gra_files)
n_shape = len(shape_files)
n_phos = len(phos_files)
n_mos2 = len(mos2_files)
n_mof = len(mof_files)
```

```python
def pdb2fp(files, cutoff, features):
    mol = []
    for i in range(len(files)):
        m = Chem.MolFromPDBFile(files[i])
        mol.append(m)
    morgan = []
    for m in mol:
        morgan.append(AllChem.GetMorganFingerprintAsBitVect(m, 
                        cutoff, useFeatures=True, nBits = features))
    return morgan
```

```python
gra_fp = pdb2fp(gra_files, cutoff_dist, n_features)
shape_fp = pdb2fp(shape_files, cutoff_dist, n_features) 
phos_fp = pdb2fp(phos_files, cutoff_dist, n_features)
mos2_fp = pdb2fp(mos2_files, cutoff_dist, n_features)
mof_fp = pdb2fp(mof_files, cutoff_dist, n_features)
```

```python
def fp2matrix(fps):
    matrices = []
    for i in range(len(fps)):
        array = np.zeros((1,));
        DataStructs.ConvertToNumpyArray(fps[i], array);
        matrices.append(array)
    matrices = np.asarray(matrices)
    return matrices
```

```python
gra_data = fp2matrix(gra_fp)
shape_data = fp2matrix(shape_fp)
phos_data = fp2matrix(phos_fp)
mos2_data = fp2matrix(mos2_fp)
mof_data = fp2matrix(mof_fp)
all_data = np.concatenate((gra_data,shape_data,
                           phos_data,mos2_data,mof_data))
```

Visualize with normal t-SNE

```python
def normal_tsne(data, perplexity = 10):
    start_shape = n_gra
    start_phos = start_shape+n_shape
    start_mos2 = start_phos+n_phos
    fig = plt.figure(figsize = [6, 4], dpi = 200)
    #########################################
    # TSNE section
    tsne = TSNE(n_components = 2,perplexity = perplexity)
    points = tsne.fit_transform(data)
    #########################################
    # Plot section
    plt.scatter(points[:n_gra,0], points[:n_gra,1], 
                label = 'Graphne')
    plt.scatter(points[start_shape:start_phos,0], 
                points[start_shape:start_phos,1],
               label = 'Weird shape Graphene')
    plt.scatter(points[start_phos:start_mos2,0], 
                points[start_phos:start_mos2,1],
               label = 'Phosphorene')
    plt.scatter(points[start_mos2:-n_mof,0], 
                points[start_mos2:-n_mof,1],
               label = r'MoS$_2$')
    plt.scatter(points[-n_mof:,0], points[-n_mof:,1],
               label = 'MOF')
    #########################################
    # Plot beautifying section
    ax = gca()
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2)
    plt.tick_params(direction = 'in',width = 1.5)
    plt.xticks([],[''])
    plt.yticks([],[''])
    plt.legend(frameon = False, loc = 'best', fontsize = 8)
```

```python
normal_tsne(all_data)
```

Compare with PCA result

```python
def pca_result(data):
    start_shape = n_gra
    start_phos = start_shape+n_shape
    start_mos2 = start_phos+n_phos
    fig = plt.figure(figsize = [6, 4], dpi = 200)
    #########################################
    # PCA section
    pca = PCA(n_components = 2)
    pca_data = pca.fit_transform(data)
    #########################################
    # Plot section
    plt.scatter(pca_data[:n_gra,0], pca_data[:n_gra,1],
               label = 'Graphne')
    plt.scatter(pca_data[start_shape:start_phos,0], 
                pca_data[start_shape:start_phos,1],
               label = 'Weird shape Graphene')
    plt.scatter(pca_data[start_phos:start_mos2,0], 
                pca_data[start_phos:start_mos2,1],
               label = 'Phosphorene')
    plt.scatter(pca_data[start_mos2:-n_mof,0], 
                pca_data[start_mos2:-n_mof,1],
               label = r'MoS$_2$')
    plt.scatter(pca_data[-n_mof:,0], pca_data[-n_mof:,1],
                label = 'MOF')
    #########################################
    # Plot beautifying section
    ax = gca()
    for axis in ['top','bottom','left','right']:
        ax.spines[axis].set_linewidth(2)
    plt.tick_params(direction = 'in',width = 1.5)
    plt.xticks([],[''])
    plt.yticks([],[''])
    plt.legend(frameon = False, loc = 'best', fontsize = 10)
```
